```python
%pylab inline
import pandas as pd
import seaborn
#SEABORN CREATES PLOTS AND PANDAS FOR VISUALISATION
```

```python
data = pd.read_csv('Desktop/uber-raw-data-apr14.csv')
data.head()
```

# Here we have converted date time into pandas date time

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
data.tail()
def get_weekday(dt):
    return dt.weekday()
data['weekday']= data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour
data['hour']= data['Date/Time'].map(get_hour)
def get_dom(dt):
    return dt.day
data['dom'] = data['Date/Time'].map(get_dom)
data.head(50)
```

# THE WEEKDAY AND HOUR OF RIDES ARE RETURNED BY DEFINING NEW COLUMNS 

# analysing the Date of the month(DOM) by using histogram

```python
hist(data.dom, bins=30 ,rwidth= 0.8 )
xlabel('Day of the Month')
ylabel('Frequency')
title('Frequency vs DOM- Uber-apr 2014')
```

#Using GROUPBY function to find frequency at each day of the month

```python
#for k, rows in data.groupby('dom'):
 #   print((k, len(rows)))
```

```python
#grouped the data in DOM by defining a new variable count_rows and by_date
```

```python
def count_rows(rows):
    return len(rows)
by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
#plotted a chart
```

```python
plot(by_date)
xlabel('Date of month')
ylabel('Frequency')
title(' Plot of Frequency vs Date of Month - uber - apr 2014')
```

```python
#sorting values of DOM 
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
#PLOTTNG BAR GRAPH OF SORTED VALUES 
```

```python
bar(range(1, 31), by_date_sorted)
xticks(range(1,31), by_date_sorted.index)
;
```
